- ### Magic Funtions

1. Object Creation & Destruction:
    - __new__(cls, ...) :- Controls the creation of a new instance (before __init__).
    - __init__(self, ...) :- Initializes an instance (like a constructor).
    - __del__(self) :- Called when an object is about to be destroyed (destructor).

2. String Representation:
    - __str__(self)	Returns a human-readable string (used by print(obj)).
    - __repr__(self) :- Returns an official string representation (repr(obj)).
    - __format__(self, format_spec) :- Custom string formatting using format(obj).
    - __bytes__(self) :- Converts the object to bytes using bytes(obj).

3. Comparison Operators:
    - __eq__(self, other) :- Implements == (equal).
    - __ne__(self, other) :- Implements != (not equal).
    - __lt__(self, other) :- Implements < (less than).
    - __le__(self, other) :- Implements <= (less than or equal).
    - __gt__(self, other) :- Implements > (greater than).
    - __ge__(self, other) :- Implements >= (greater than or equal).

4. Arithmetic Operators:
    - __add__(self, other) :- Implements +.
    - __sub__(self, other) :- Implements -.
    - __mul__(self, other) :- Implements *.
    - __truediv__(self, other) :- Implements /.
    - __floordiv__(self, other) :- Implements //.
    - __mod__(self, other) :- Implements %.
    - __pow__(self, other) :- Implements **.
    - __radd__(self, other) :- Implements right-hand + (other + self).
    - __rsub__(self, other) :- Implements right-hand -.
    - __rmul__(self, other) :- Implements right-hand *.
    - __rtruediv__(self, other) :- Implements right-hand /.

5. Bitwise Operators:
    - __and__(self, other) :- Implements & (bitwise AND).
    - __or__(self, other) :- Implements `
    - __xor__(self, other) :- Implements ^ (bitwise XOR).
    - __lshift__(self, other) :- Implements << (left shift).
    - __rshift__(self, other) :- Implements >> (right shift).
    - __invert__(self) :- Implements ~(bitwise NOT).

6. Assignment Operators:
    - __iadd__(self, other) :- Implements +=.
    - __isub__(self, other) :- Implements -=.
    - __imul__(self, other) :- Implements *=.
    - __itruediv__(self, other) :- Implements /=.
    - __ifloordiv__(self, other) :- Implements //=.
    - __imod__(self, other) :- Implements %=.
    - __ipow__(self, other) :- Implements **=.

7. Type Conversion:
    - __int__(self) :- Converts object to int().
    - __float__(self) :- Converts object to float().
    - __complex__(self) :- Converts object to complex().
    - __bool__(self) :- Converts object to bool().
    - __hash__(self) :- Returns a hash value (used in sets/dictionaries).

8. Container & Collection Methods:
    - __len__(self) :- Implements len(obj).
    - __getitem__(self, key) :- Implements obj[key].
    - __setitem__(self, key, value) :- Implements obj[key] = value.
    - __delitem__(self, key) :- Implements del obj[key].
    - __contains__(self, item) :- Implements item in obj.
    - __iter__(self) :- Implements iteration (for x in obj).
    - __next__(self) :- Implements next(obj) (for iterators).

9. Callable & Context Manager Methods:
    - __call__(self, ...) :- Allows an instance to be called like a function.
    - __enter__(self) :- Implements with obj as ....
    - __exit__(self, exc_type, exc_val, exc_tb) :- Implements cleanup after with block.

10. Custom Attribute Access:
    - __getattr__(self, name) :- Called when an attribute is not found.
    - __setattr__(self, name, value) :- Called when setting an attribute.
    - __delattr__(self, name) :- Called when deleting an attribute.
    - __dir__(self) :- Returns a list of available attributes.

11. Memory Management:
    - __sizeof__(self)	Returns the size of the object in bytes.
